```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statistics import mean
import random
```

```python
## Generated below is a sample of 7 observation from group1 and 6 from group2
## Sample A is taken from a mean of 100 and std deviation of 20
## Sample b is taken from a mean of 50 and std deviation of 15
```

```python
random.seed(20) #for results to be recreated
N1 = 7 #number of samples to be taken from Customer group 1
N2 = 6 #number of samples to be taken from Customer group 2

a = [random.gauss(100,20) for x in range(N1)] #take N samples from population A
b = [random.gauss(50,15) for x in range(N2)] #take N samples from population B
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(a, shade=True)
sns.kdeplot(b, shade=True)
plt.title("Independent  Customers Sample T-Test")
```

```python
tStat, pValue = stats.ttest_ind(a, b, equal_var = False) #run independent sample T-Test
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat)) #print the P-Value and the T-Statistic
```

```python
## The P value is lower than 5% so it should not be accepted. We used the assuption as regards to the variance
```
